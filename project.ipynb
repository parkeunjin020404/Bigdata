```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

def preprocess_health_data(file_path):
    # 데이터 로드
    data = pd.read_csv(file_path, encoding='euc-kr')

    # 불필요한 컬럼 제거
    drop_columns = ['총콜레스테롤', '트리글리세라이드', 'HDL콜레스테롤', 'LDL콜레스테롤',
                    '시력(좌)', '시력(우)', '치아우식증유무', '치석', '가입자일련번호']
    data = data.drop(columns=drop_columns)

    # BMI 계산 함수
    def calculate_bmi(row):
        height_m = row['신장(5cm단위)'] / 100  # 신장을 cm에서 m로 변환
        weight_kg = row['체중(5kg단위)']  # 체중은 kg 단위
        return weight_kg / (height_m ** 2)

    # BMI 컬럼 추가
    data['BMI'] = data.apply(calculate_bmi, axis=1)
    data['비만'] = data['BMI'] >= 25  # 비만 여부 (True/False)

    # 연령대 코드와 연령 범위를 매핑한 딕셔너리 생성
    age_range_map = {
        5: '20-24세',
        6: '25-29세',
        7: '30-34세',
        8: '35-39세',
        9: '40-44세',
        10: '45-49세',
        11: '50-54세',
        12: '55-59세',
        13: '60-64세',
        14: '65-69세',
        15: '70-74세',
        16: '75-79세',
        17: '80-84세',
        18: '85세 이상'
    }
    # 연령범위 컬럼 추가
    data['연령범위'] = data['연령대코드(5세단위)'].map(age_range_map)

    # 복부비만 계산 함수
    def abdominal_obesity(row):
        if row['성별'] == 1:  # 남성
            return row['허리둘레'] >= 90
        elif row['성별'] == 2:  # 여성
            return row['허리둘레'] >= 85
        else:
            return False

    # 복부비만 컬럼 추가
    data['복부비만'] = data.apply(abdominal_obesity, axis=1)

    return data



# 각 연도별 파일 경로 지정
#file_path_2017 = '/content/drive/MyDrive/Colab Notebooks/bigdata/healthcare2017.csv'
#file_path_2018 = '/content/drive/MyDrive/Colab Notebooks/bigdata/healthcare2018.csv'
#file_path_2019 = '/content/drive/MyDrive/Colab Notebooks/bigdata/healthcare2019.csv'
file_path_2020 = '/content/drive/MyDrive/Colab Notebooks/bigdata/healthcare2020.CSV'
#file_path_2021 = '/content/drive/MyDrive/Colab Notebooks/bigdata/healthcare2021.csv'
#file_path_2022 = '/content/drive/MyDrive/Colab Notebooks/bigdata/healthcare2022.csv'
#file_path_2023 = '/content/drive/MyDrive/Colab Notebooks/bigdata/healthcare2023.csv'

# 데이터 처리
#data2017 = preprocess_health_data(file_path_2017)
#data2018 = preprocess_health_data(file_path_2018)
#data2019 = preprocess_health_data(file_path_2019)
data2020 = preprocess_health_data(file_path_2020)
#data2021 = preprocess_health_data(file_path_2021)
#data2022 = preprocess_health_data(file_path_2022)
#data2023 = preprocess_health_data(file_path_2023)

# 성별 1 (남성) 데이터
#male_data2017 = data2017[data2017['성별'] == 1]
#male_data2018 = data2018[data2018['성별'] == 1]
#male_data2019 = data2019[data2019['성별'] == 1]
male_data2020 = data2020[data2020['성별'] == 1]
#male_data2021 = data2021[data2021['성별'] == 1]
#male_data2022 = data2022[data2022['성별'] == 1]
#male_data2023 = data2023[data2023['성별'] == 1]

# 성별 2 (여성) 데이터
#female_data2017 = data2017[data2017['성별'] == 2]
#female_data2018 = data2018[data2018['성별'] == 2]
#female_data2019 = data2019[data2019['성별'] == 2]
female_data2020 = data2020[data2020['성별'] == 2]
#female_data2021 = data2021[data2021['성별'] == 2]
#female_data2022 = data2022[data2022['성별'] == 2]
#female_data2023 = data2023[data2023['성별'] == 2]

def calculate_male_stats(data):
    bmi_mean = data['BMI'].mean()
    obesity_rate = data['비만'].mean()
    abdominal_obesity_rate = data['복부비만'].mean()
    age_distribution = data['연령범위'].value_counts()

    return {
        'BMI 평균': bmi_mean,
        '비만 비율': obesity_rate,
        '복부비만 비율': abdominal_obesity_rate,
        '연령범위': age_distribution
    }

#male_stats_2017 = calculate_male_stats(male_data2017)
#male_stats_2018 = calculate_male_stats(male_data2018)
#male_stats_2019 = calculate_male_stats(male_data2019)
male_stats_2020 = calculate_male_stats(male_data2020)
#male_stats_2021 = calculate_male_stats(male_data2021)
#male_stats_2022 = calculate_male_stats(male_data2022)
#male_stats_2023 = calculate_male_stats(male_data2023)

def calculate_female_stats(data):
    bmi_mean = data['BMI'].mean()
    obesity_rate = data['비만'].mean()
    abdominal_obesity_rate = data['복부비만'].mean()
    age_distribution = data['연령범위'].value_counts()

    return {
        'BMI 평균': bmi_mean,
        '비만 비율': obesity_rate,
        '복부비만 비율': abdominal_obesity_rate,
        '연령범위': age_distribution
    }

#female_stats_2017 = calculate_female_stats(female_data2017)
#female_stats_2018 = calculate_female_stats(female_data2018)
#female_stats_2019 = calculate_female_stats(female_data2019)
female_stats_2020 = calculate_female_stats(female_data2020)
#female_stats_2021 = calculate_female_stats(female_data2021)
#female_stats_2022 = calculate_female_stats(female_data2022)
#female_stats_2023 = calculate_female_stats(female_data2023)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')  # NanumGothic 폰트 사용
plt.rc('axes', unicode_minus=False)  # 마이너스 기호가 깨지는 문제 해결
# 통계 정보 추출
male_stats = {
    'BMI 평균': male_stats_2020['BMI 평균'],
    '비만 비율': male_stats_2020['비만 비율'],
    '복부비만 비율': male_stats_2020['복부비만 비율']
}

female_stats = {
    'BMI 평균': female_stats_2020['BMI 평균'],
    '비만 비율': female_stats_2020['비만 비율'],
    '복부비만 비율': female_stats_2020['복부비만 비율']
}

# 데이터프레임 생성
stats_comparison = pd.DataFrame({
    '성별': ['남성', '여성'],
    #'BMI 평균': [male_stats['BMI 평균'], female_stats['BMI 평균']],
    '비만 비율': [male_stats['비만 비율'], female_stats['비만 비율']],
    '복부비만 비율': [male_stats['복부비만 비율'], female_stats['복부비만 비율']]
})

print(stats_comparison)

# 데이터 변환
stats_melted = stats_comparison.melt(id_vars='성별', var_name='통계', value_name='값')

# 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x='통계', y='값', hue='성별', data=stats_melted, palette='Set1')
plt.title('2020년 성별에 따른 건강 통계 비교')
plt.xlabel('통계 항목')
plt.ylabel('값')
plt.ylim(0, 1)  # 비율이 0에서 1 사이이므로 y축 범위 설정
plt.grid(axis='y')  # y축 그리드 추가
plt.legend(title='성별')
# 한글 폰트 설정

plt.show()
```
